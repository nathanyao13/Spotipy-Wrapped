## User Generation

```python
def generate_user_item(song_df, num_users=500, min_tracks=10, max_tracks=50, max_plays=20):
    user_item_data = []
    track_ids = song_df['id'].dropna().unique()

    for user_index in range(1, num_users + 1):
        user_id = user_index
        n_tracks = random.randint(min_tracks, max_tracks)
        listened_tracks = np.random.choice(track_ids, size=n_tracks, replace=False)

        for track_id in listened_tracks:
            play_count = random.randint(1, max_plays)
            user_item_data.append({
                'user_id': user_id,
                'track_id': track_id,
                'play_count': play_count
            })

    return pd.DataFrame(user_item_data)
```

```python
user_item_df = generate_user_item(df, num_users=1000)

user_item_df.to_csv("user_item.csv", index=False)
```
