# EDA and User Generation

```python
import pandas as pd
import numpy as np
import random
```

## Music Metadata

```python
df = pd.read_csv("metadata.csv", header = 0)
```

## User Generation

```python
track_ids = df['id'].tolist()
num_tracks = len(track_ids)

num_users = 100          
likes_per_user = 100      

user_ids = [i for i in range(num_users)]

user_item_matrix = pd.DataFrame(0, index=user_ids, columns=track_ids)

for user in user_ids:
    liked_tracks = random.sample(track_ids, likes_per_user)
    user_item_matrix.loc[user, liked_tracks] = 1

user_item_matrix.to_csv("user_item.csv", index=False)
```
